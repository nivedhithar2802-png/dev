```python
# Import Libraries
import pandas as pd
import matplotlib.pyplot as plt

# Read Dataset
df = pd.read_csv("titanic.csv")

# Display First Five Rows
print(df.head())

# Display Dataset Information
print(df.info())

# Identify Missing Values
print(df.isnull().sum())

# Histogram Before Cleaning
plt.figure(figsize=(7,5))
plt.hist(df["Age"], bins=15, edgecolor="black")
plt.title("Age Distribution Before Cleaning")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()

# Handle Missing Values
# Replace missing Age values using the median
df["Age"] = df["Age"].fillna(df["Age"].median())

# Verify Missing Values
print(df.isnull().sum())

# Histogram After Cleaning
plt.figure(figsize=(7,5))
plt.hist(df["Age"], bins=15, edgecolor="black")
plt.title("Age Distribution After Cleaning")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
# Import Libraries
import pandas as pd
import matplotlib.pyplot as plt

# Read Dataset
df = pd.read_csv("titanic.csv")

# Display First Five Rows
print(df.head())

# Display Dataset Information
print(df.info())

# Scatter Plot Before Removing Outliers
plt.figure(figsize=(8,5))
plt.scatter(df.index, df["Fare"])
plt.title("Fare Before Removing Outliers")
plt.xlabel("Passenger Number")
plt.ylabel("Fare")
plt.grid(True)
plt.show()

# Calculate IQR
Q1 = df["Fare"].quantile(0.25)
Q3 = df["Fare"].quantile(0.75)
IQR = Q3 - Q1

# Define Lower and Upper Limits
lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR

# Remove Outliers
df_clean = df[(df["Fare"] >= lower_limit) & (df["Fare"] <= upper_limit)]

# Scatter Plot After Removing Outliers
plt.figure(figsize=(8,5))
plt.scatter(df_clean.index, df_clean["Fare"])
plt.title("Fare After Removing Outliers")
plt.xlabel("Passenger Number")
plt.ylabel("Fare")
plt.grid(True)
plt.show()

# Display Shape Before and After
print("Original Shape:", df.shape)
print("Shape After Removing Outliers:", df_clean.shape)
```
