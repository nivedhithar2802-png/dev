```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("Iris.csv")
df.head()
df.info()
```

```python
plt.figure(figsize=(8,5))
plt.plot(df["Id"], df["SepalLengthCm"], marker='o')

plt.title("Sepal Length of Iris Flowers")
plt.xlabel("Flower ID")
plt.ylabel("Sepal Length (cm)")
plt.grid(True)

plt.show()
```

```python


# Scatter Plot
plt.figure(figsize=(8,5))
plt.scatter(df["SepalLengthCm"], df["SepalWidthCm"])

plt.title("Sepal Length vs Sepal Width")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
plt.grid(True)

plt.show()
```

```python
species_count = df["Species"].value_counts()

# Bar Chart
plt.figure(figsize=(8,5))
plt.bar(species_count.index, species_count.values)

plt.title("Count of Iris Flower Species")
plt.xlabel("Species")
plt.ylabel("Count")
plt.grid(axis='y')

plt.show()
```

```python
species_count = df["Species"].value_counts()

# Pie Chart
plt.figure(figsize=(7,7))
plt.pie(species_count,
        labels=species_count.index,
        autopct='%1.1f%%',
        startangle=90)

plt.title("Distribution of Iris Flower Species")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(df["SepalLengthCm"], bins=10, edgecolor='black')

plt.title("Histogram of Sepal Length")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Frequency")
plt.grid(True)

plt.show()
```
